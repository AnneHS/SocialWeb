```python
import requests
with open('week1(24_02-01_3)dominos_data.csv') as csvfile:
    dialect = csv.Sniffer().sniff(csvfile.read(14734))

with open('week1(24_02-01_3)pizzahut_data.csv') as csvfile_:
    dialect_ = csv.Sniffer().sniff(csvfile_.read(14734))


public_1 = pd.read_csv('week1(24_02-01_3)dominos_data.csv', engine='python', dialect=dialect, error_bad_lines=False)
public_2 = pd.read_csv('week1(24_02-01_3)pizzahut_data.csv', engine='python', dialect=dialect_, error_bad_lines=False)
```

```python
time_all = public_1['time']
time_all_p = public_2['time']
```

# Plot the tweet counts according to the time (morning,afternoon,night)

```python
morning = 0
afternoon = 0
night = 0


#for time in time_all:
    #print(time[10:][:3])

for time in time_all:
    try:
        if int(float(time[10:][:3]))>6 and int(float(time[10:][:3]))<12:
            morning += 1
        elif int(float(time[10:][:3]))>12 and int(float(time[10:][:3]))<18:
            afternoon += 1
        else:
            night += 1
    except ValueError:
        print(time)
    
count = dict()
count['morning'] = morning
count['afternoon'] = afternoon
count['night'] = night
print(count)
```

```python
print(morning+afternoon+night)
```

```python
plt.figure(figsize=(8,6),dpi=100)
plt.plot(list(count.keys()),list(count.values()),'k')
```
